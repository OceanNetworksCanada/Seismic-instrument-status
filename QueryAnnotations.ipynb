```python
resourceId='11057' # device Ids

# query to include topology
url = 'https://data.oceannetworks.ca/AnnotationServiceV3?annotationSources=3%2C5&resourceTypeId=2&includeTopology=true&resourceId='+resourceId+'&fields=%7B%7D&levelIndexes=0%2C1%2C2%2C3%2C4%2C5&pageSize=1000&pageNum=1'

# query to not include topology
url2 = 'https://data.oceannetworks.ca/AnnotationServiceV3?annotationSources=3%2C5&resourceTypeId=2&includeTopology=false&resourceId='+resourceId+'&fields=%7B%7D&levelIndexes=0%2C1%2C2%2C3%2C4%2C5&pageSize=1000&pageNum=1'
```

```python
import requests
```

```python
# gt the URL into a json
r = requests.get(url2)

len(r.json()['payload']['naiveAnnotationList'])
```

```python
# find the annoations
r.json()['payload']['naiveAnnotationList']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(r.json()['payload']['naiveAnnotationList'])
```

```python
df.set_index(pd.to_datetime(df['startDate'],utc=False),inplace=True)
df.index = df.index.tz_localize(None)
df
```

```python
df.sort_index(ascending=False).iloc[0]['annotationSummary']
```
